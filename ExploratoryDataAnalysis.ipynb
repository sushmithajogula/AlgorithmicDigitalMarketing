## Exploratory Data Analysis (EDA)

In simple terms, Exploratory Data Analysis (EDA) is all about trying to understand the given data much better, so that we can make some sense out of it. Exploratory Data Analysis (EDA) is the process of understanding the data sets by summarizing their main characteristics and plotting them visually. It often takes much time to explore the data but this is an important process when we arrive at modeling the data in order to apply Machine learning. As part of plotting our dataset in the EDA process , we have used the matplotlib library consisting of several kinds of plots such as Bar Plot, Pie Chart, Histograms, Treemap, Scatter plot and many more.

### Graphical Data Exploration and Analysis

#### What are the top Product Categories at Global Superstore?

```python
orders_final_df['product_category_name_english']=[str(x) for x in orders_final_df['product_category_name_english']]
categories=",".join(orders_final_df["product_category_name_english"])
categories= [x.strip() for x in categories.split(',')]

categories=Counter(categories)
top_15_categories=[x for x in categories.most_common(15)]
a,b=map(list,zip(*top_15_categories))
plt.figure(figsize=(8,8))
sns.barplot(b,a,palette="Spectral")
plt.title("Top Product Categories in Global Superstore Sales")
plt.ylabel("Types of Product Categories")
plt.xlabel("Count")
```

The top product categories in Global Superstore sales are bed_bath_table, health_beauty, and sports_leisure respectively. These product categories generate the highest amount of revenue for Global Superstore and the products from these categories are present in most number of orders made by customers. 

#### What are the preferred Payment Types amongst Global Superstore customers?

```python
orders_final_df['payment_type']=[str(x) for x in orders_final_df['payment_type']]
ptype=",".join(orders_final_df["payment_type"])
ptype= [x.strip() for x in ptype.split(',')]

ptype=Counter(ptype)
top_ptype=[x for x in ptype.most_common(5)]
a,b=map(list,zip(*top_ptype))
plt.figure(figsize=(8,8))
colors = sns.color_palette('pastel')[0:5]
plt.pie(b,labels = a, colors = colors, autopct='%.0f%%')
plt.show()
```

Here, we can clearly see that all the payments made at Global Superstore are single-handedly dominated by credit_card with almost 75% of the payment transactions happening via credit_card, making it the most preferred payment method by customers. The other payment methods used by customers are boleto, voucher, and debit_card.

#### What are the top cities with most number of orders?

```python
plt.figure(figsize = (8,8))
orders_final_df.customer_city.value_counts().nlargest(10).plot(kind='barh', color = 'yellow')
plt.title("Number of orders by cities")
plt.xlabel("Order counts")
plt.ylabel("Cities")
plt.show()
```

Here, we can see that Sao Paulo leads predominantly with the most number of orders placed from there, followed by Rio De Janeiro. We can conclude that Global Superstore and its products are very famous amongst the people of Sao Paulo. 

#### What is the approximate spending range among customers per state and city for order?

```python
temp = orders_final_df.groupby(['customer_state','customer_city'])['total_payment'].agg('mean').sort_values(ascending = False).head(20)
ax = temp.plot(figsize=(25,8), grid=True, kind = 'area', color = 'pink')
ax.set_xticks(range(len(temp)))
ax.set_xticklabels(temp.index, rotation=90, fontsize=15)
ax.set_xlabel('Order State and City', fontsize=20)
ax.set_ylabel('Order Cost', fontsize=20)
ax.set_title('Approximate spending per state and city', fontsize=20, pad = 10)
plt.show()
```

We can see that the average spending of customer per order in the city Pianco of state PB is the highest amongst all other cities and states. We can say that the customers in this city are not very price sensitive, hence, products can be attempted to be sold at a higher price than usual at Pianco. 

#### What is the distribution of orders per weekday?

```python
temp1 = orders_final_df.groupby('order_weekday')['order_id'].count().sort_values(ascending=False).reset_index()
ax = temp1.plot(kind='bar', color = 'green')
ax.set_xticks(temp1.index)
ax.set_xticklabels(temp1['order_weekday'], fontsize=10)
ax.set_ylabel('Order Count', fontsize=15)
ax.set_xlabel('Weekday', fontsize=15)
ax.set_title('Distribution of orders per weekday', fontsize=15)
plt.legend(loc='best')
plt.show()
```

We can see that Monday is the most popular day for customers to place on order at Global Superstore and this drops as we proceed along the week, with the lowest number of orders being placed on Saturday. Hence, we can say that people are more eager to buy items during the starting days of the week. 

#### Distribution of orders per weekday and hour

```python
temp5 = orders_final_df
temp5['order_date'] = pd.to_datetime(temp5['order_date'])
temp5['order_purchase_hour'] = temp5.order_date.apply(lambda x: x.hour)
temp1 = temp5.groupby(['order_weekday', 'order_purchase_hour']).nunique()['order_id'].unstack()
plt.figure(figsize=(20,8))
sns.heatmap(temp1.reindex(index = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']), cmap=sns.light_palette("seagreen", as_cmap=True), annot=True, fmt="d", linewidths=0.2)
plt.xlabel('Purchase Hour')
plt.ylabel('Day of Week')
plt.show()
```

Expanding on the above graph, we can see that early afternoon times between 10 AM and 4 PM is when the customers are most active and Global Superstore receives most number of orders, especially on the starting days of the week. 

#### What is the distribution of orders per month?

```python
temp5 = orders_final_df
temp5['order_month'] = temp5['order_month'].apply(lambda x: calendar.month_abbr[x])
temp1 = temp5.groupby('order_month')['order_id'].count().sort_values(ascending=True).reset_index()
#plt.figure(figsize = (8,8))
#temp = df.groupby('type')['rate'].mean().reset_index()
ax = temp1.plot(kind='bar', color = 'green')
ax.set_xticks(temp1.index)
ax.set_xticklabels(temp1['order_month'], fontsize=10)
ax.set_ylabel('Order Count', fontsize=15)
ax.set_xlabel('Month', fontsize=15)
ax.set_title('Distribution of orders per month', fontsize=15)
plt.legend(loc='best')
plt.show()
```

Summarizing all the order transactions, we can observe that the most number of orders are placed between the summer and fall seasons of the year, primarily in August and May respectively. Note that the number of orders drops from September and is relatively low till the end of the year; the customers are not probably inclined to place orders during the end of the year. 

#### What is the distribution of orders per year?

```python
temp5 = orders_final_df
temp5['order_year'] = temp5['order_year'].astype(str)
temp1 = temp5.groupby('order_year')['order_id'].count().reset_index()
print(temp1)
#plt.figure(figsize = (8,8))
#temp = df.groupby('type')['rate'].mean().reset_index()
ax = temp1.plot(kind='bar', color = 'green')
ax.set_xticks(temp1.index)
ax.set_xticklabels(temp1['order_year'], fontsize=10)
ax.set_ylabel('Order Count', fontsize=15)
ax.set_xlabel('Year', fontsize=15)
ax.set_title('Distribution of orders per year', fontsize=15)
plt.legend(loc='best')
plt.show()
```

#### Correlation matrix

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
sns.heatmap(orders_final_df.corr(), mask=np.triu(orders_final_df.corr()), annot=True)
plt.title('Correlation of features', size=20)
plt.show()
```

A correlation matrix is a table containing correlation coefficients between variables where each cell in the table represents the correlation between two variables. The value of correlation coefficients lies between -1 and 1. This matrix is used to summarize data, as a diagnostic for advanced analyses and as an input into a more advanced analysis. 

In this case, we can see that the columns total_payment and order_item_id are most correlated, and the columns order_item_id and review_score are least correlated.
